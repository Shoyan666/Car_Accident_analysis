```python
import numpy as np 
import pandas as pd 
# importing libraries
```

```python
# loading dataset
road = pd.read_csv("Dataset_pandas_assign.csv")
road.head(5)
```

```python
# deleting first row which was the unnecessary with indexes
road = road.iloc[1:,:]
road.head(5)
```

```python
road.groupby('Пол', as_index = False)['Возраст, полных лет'].mean().round()
# average age of males grater on 4
```

```python
# by checking one of the determinants of accident we get rid of all the non-accident rows
road_accidents = road[~road['Сумма убытка'].isna()]
road_accidents
```

```python
gender_participation = road_accidents[['Пол', 'Unique number']].groupby('Пол').count()
gender_participation
```

```python
print('car accidents with female', gender_participation.iloc[0]/len(road_accidents))
print('car accidents with male', gender_participation.iloc[1]/len(road_accidents))
# here we can observe that males participate in car accidents almost in 81 percent of cases witch is huge point
# this task was discussed with Adal Abilbekov, thanks to him for providing an advice
```

### Task 3 In which year the most significant number of vehicles were built? Please provide the list of top-10 years sorting based on the number of vehicles (descending).

```python
# creating new dataframe to get only years from date
road_new = road.copy()
road_new['year'] = pd.DatetimeIndex(road_new['Год выпуска']).year  
road_new.head(5)
```

```python
# sorting by count of vehicles 
road_sorted = road_new.groupby('year').count().sort_values('Unique number', ascending = False)
road_sorted.rename(columns = {'Unique number':'count'}, inplace = True)
road_sorted.iloc[0:10,0:1]
```

### Task 5 What is the most popular vehicle-make company in any 5 cities? (choose the cities you like most) 


```python
# for Almaty city
road_almaty = road[road['Город'] == "Алматы"][['Город', 'Модель', 'Unique number']].groupby(['Модель']).count().sort_values('Unique number', ascending=False)
road_almaty.head(1)
# we found that the most popular vehicle-make company is Toyota
```

```python
# for Nur-Sultan city
road_nursultan = road[road['Город'] == "Нур-Султан"][['Город', 'Модель', 'Unique number']].groupby(['Модель']).count().sort_values('Unique number', ascending=False)
road_nursultan.head(1)
# we found that the most popular vehicle-make company is Toyota
```

```python
# for Shymkent city
road_shymkent = road[road['Город'] == "Шымкент"][['Город', 'Модель', 'Unique number']].groupby(['Модель']).count().sort_values('Unique number', ascending=False)
road_shymkent.head(1)
# we found that the most popular vehicle-make company is Volkswagen
```

```python
# for Atyrau city
road_atyrau = road[road['Город'] == "Атырау"][['Город', 'Модель', 'Unique number']].groupby(['Модель']).count().sort_values('Unique number', ascending=False)
road_atyrau.head(1)
# we found that the most popular vehicle-make company is ВАЗ
```

```python
# for Oskemen city
road_oskemen = road[road['Город'] == "Усть-Каменогорск"][['Город', 'Модель', 'Unique number']].groupby(['Модель']).count().sort_values('Unique number', ascending=False)
road_oskemen.head(1)
# we found that the most popular vehicle-make company is Toyota
```

```python
# there is another way to do this task also this method is much easier than the previous one
top5_cities = ['Алматы', 'Нур-Султан', 'Шымкент', 'Атырау', 'Усть-Каменогорск'] 
city_most_popular_vehicle_makers = road[road['Город'].isin(top5_cities)][['Город', 'Модель', 'Unique number']].groupby(['Город', 'Модель']).count().sort_values('Unique number', ascending=False).reset_index().drop_duplicates('Город', keep='first') 
city_most_popular_vehicle_makers
```
